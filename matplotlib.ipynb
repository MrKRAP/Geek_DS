```python
t = np.linspace(0,10,51)
f = np.cos(t)
plt.plot(t,f)
plt.title('График f(t)',color = 'green')
plt.ylabel('Значение f')
plt.xlabel('Значение t')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
x = np.linspace(-3,3,51)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

# create cols and rows
fig,ax = plt.subplots(2,2)
ax1,ax2,ax3,ax4 = ax.flatten()
# Изменил значения так как некрасиво)))
fig.set_size_inches(10,10)
fig.subplots_adjust(0.1,0.1)
# ax[0].axis(x,y1)
# ax[0].plot(x,y1)
ax1.plot(x,y1)
ax1.set_title('График y1')
ax1.set_xlim([-5,5])
ax2.plot(x,y2)
ax2.set_title('График y2')
ax3.plot(x,y3)
ax3.set_title('График y3')
ax4.plot(x,y4)
ax4.set_title('График y4')

# plt.subplot(141)
# plt.plot(x,y1)
# plt.subplot(142)
# plt.plot(x,y2)
# plt.subplot(143)
# plt.plot(x,y3)
```
